## ***Importing Necessary Libraries***

```python
import matplotlib.pyplot as plt
import numpy as np
```

## ***Creating the Dataset***

```python
x = [1,2,3,4] # input
y = [2,4,6,8] # output
```

## ***Visualizing the Dataset***

```python
plt.scatter(x,y)
plt.show()
```

## ***Defining the Cost Function - Mean Squared Error***

image.png

```python
def computeCost(x, y, slope):
    inner = np.power(((x @ slope.T) - y), 2) # @ means matrix multiplication of arrays
    return np.sum(inner) / (2 * len(x))
```

## ***Defining Gradient Descent***

```python
def gradientDescent(x, y, slope, learning_rate, iters):
    for i in range(iters):
        cost = computeCost(x, y, slope) # Error
        slope = slope - (learning_rate/len(x)) * np.sum((x @ slope.T - y), axis=0) # Slope Updation
        if(i%100==0):
          print(i, cost, slope)
    return(slope, cost)
```

## ***Running the Algorithm***

```python
learning_rate = 0.0001
iters = 500
initial_slope = np.array([[3, 3, 3, 3]])
print('initial_slope : ',initial_slope)
slope, cost = gradientDescent(x, y, initial_slope, learning_rate, iters)  
print(iters ,cost, slope)
```

## ***Visualizing the Result***

```python
plt.scatter(x, y) # input and output
axes = plt.gca()
x_vals = np.array(axes.get_xlim()) 
y_vals = slope[0][0] * x_vals # the line equation
plt.plot(x_vals, y_vals, '--',c='r')
plt.show()
```
